```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf
from src import DRCClass
```

```python
S_t = 9950
t = 0
T = 60
K = 10000 * (1 - 0.02)
r = 0.001
volatility = 0.10
results = []
for S_t in [10000*(1 + i/100) for i in range(-2, 10)]:
    BS = DRCClass.BlackScholes(S_t, t, T, K, r, volatility)
    d = BS.call_greeks()
    results.append([d["premium"], d["delta"], d["vega"]])
```

```python
results = np.array(results).T
```

```python
plt.figure(figsize=(8,3))
plt.plot(results[0])
plt.plot(results[2])
plt.figure(figsize=(8,3))
plt.plot(results[1])
```

```python
plt.figure(figsize=(8,3))
plt.plot([10000*(1 + i/100) for i in range(-2, 10)],results[1]/results[0]*(np.array([10000*(1 + i/100) for i in range(-2, 10)])-K))
plt.figure(figsize=(8,3))
plt.plot(results[1]/results[0]*(S_t))
```
