```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf
from src import DRCClass
```

```python
S_t = 10000
T = 250
K1 = S_t * (1 - 0.10)
K2 = S_t * (1 - 0.05)
K3 = S_t * (1 - 0.0)
K4 = S_t * (1 + 0.05)
K5 = S_t * (1 + 0.10)

r = 0.003
volatility = 0.20
premiums = np.zeros((250,5))
deltas = np.zeros((250,5))
for t in range(250):
    for idx, K in enumerate([K1,K2,K3,K4,K5]):
        BS = DRCClass.BlackScholes(S_t, t/250, T/250, K, r, volatility)
        d = BS.call_greeks()
        premiums[t, idx] = d["premium"]
        deltas[t, idx] = d["delta"]
```

```python
plt.figure(figsize=(12,6))
for i in range(5):
    plt.plot(range(250), premiums.T[i], color="blue", alpha=0.2*(i+1))

plt.figure(figsize=(12,6))
for i in range(5):
    plt.plot(range(250), deltas.T[i], color="blue", alpha=0.2*(i+1))
```
